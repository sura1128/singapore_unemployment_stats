```python
temp_dict = {"Year": [],
             "Age": [],
             "Occupation": [],
             "Employed": []}
years = df.keys().tolist()
years.remove("Occupation")
years.remove("Age")
print (years)
```

```python
# now we have the keys
for year in years:
  temp_dict["Year"].extend([year]*99)
  temp_dict["Age"].extend(df["Age"])
  temp_dict["Occupation"].extend(df["Occupation"])
  temp_dict["Employed"].extend(df[year])


shaped_df = pd.DataFrame.from_dict(temp_dict)
shaped_df.head(100)
```

```python
shaped_df.info()
```

```python
shaped_df.to_csv("/content/M182171_Employed_Residents_Age.xlsx - employed_age_SHAPED.csv", encoding='utf-8', index=False)
```
